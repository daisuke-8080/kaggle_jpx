```python
train_data = pd.read_csv('/kaggle/input/jpx-tokyo-stock-exchange-prediction/train_files/stock_prices.csv')
train_length = len(train_data[(train_data['SecuritiesCode'] == 1301)])
Unique = np.unique(train_data['SecuritiesCode'].values)
```

```python
df_preds = pd.read_csv('../input/close-predicted/close_predicted.csv')
```

```python
del df_preds['Unnamed: 0']
```

```python
Unique = np.unique(train_data['SecuritiesCode'].values)
Unique = list(Unique)
Unique = [int(i) for i in Unique]
```

```python
for i in range(365):
    df_preds = df_preds.rename(columns={f'1{202+i}':f'{i+1}'})
df_preds
```

```python
df_preds = df_preds.reset_index(drop=True)
Unique_dict= {}
for i, unique in enumerate(Unique):
    Unique_dict[i] = unique
df_preds = df_preds.rename(index=Unique_dict)
```

```python
df_null_index = pd.DataFrame([df_preds.isnull().sum(axis=1)])
df_null_index = df_null_index.transpose().rename(columns={0:'null_exist'})
df_null_exist = df_null_index[(df_null_index['null_exist'] != 0)]
```

```python
null_list = [int(i) for i in list(df_null_exist.index)]
len(null_list)
```

```python
df_preds_null = pd.read_csv('../input/close-null-predected-new/close_null_predicted_new.csv')
```

```python
del df_preds_null['Unnamed: 0']
```

```python
for i in range(365):
    df_preds_null = df_preds_null.rename(columns={f'1{202+i}':f'{i+1}'})
df_preds_null
```

```python
for i,unique in enumerate(null_list):

    df_preds_null = df_preds_null.rename(index={i:unique})
```

```python
df_preds = df_preds.drop(null_list)
```

```python
df_all = pd.concat((df_preds, df_preds_null), axis=0).sort_index()
df_all
```

```python
for code in df_all.index:
    for i in range(363):
        i = i+1
        df_all.loc[code,f'{i}'] = (df_all.loc[code,f'{i + 2}'] - df_all.loc[code,f'{i + 1}']) / df_all.loc[code,f'{i + 1}']

df_all
```

```python
del df_all['364']
del df_all['365']
```

```python
df_all = df_all.rank(method='first',ascending=False).astype(int) -1
```

```python
df_all.reset_index(inplace=True)
```

```python
import jpx_tokyo_market_prediction
env = jpx_tokyo_market_prediction.make_env()   # initialize the environment
iter_test = env.iter_test()    # an iterator which loops over the test files
i = '1'
for (prices, options, financials, trades, secondary_prices, sample_prediction) in iter_test:
    del sample_prediction['Rank']
    sample_prediction['Rank'] = df_all[[i]]
    i = str(int(i) + 1)
    print(sample_prediction)
    env.predict(sample_prediction)
```
